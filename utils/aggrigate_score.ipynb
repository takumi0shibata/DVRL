```python
import numpy as np
import pandas as pd

pd.options.display.float_format = '{:,.3f}'.format
```

```python
outputs_path = 'outputs/'
experiments_name = 'fastPAES'
seeds = [12, 22, 32, 42, 52]
prompts = [i for i in range(1, 9)]
```

```python
output_df = pd.DataFrame(columns=['qwk', 'lwk', 'corr', 'rmse', 'mae', 'seed', 'prompt'])
for seed in seeds:
    for prompt in prompts:
        filename = f'../{outputs_path}{experiments_name}/{seed}/best_metrics_prompt{prompt}.csv'
        df = pd.read_csv(filename)
        df['seed'] = seed
        df['prompt'] = prompt
        output_df = pd.concat([output_df, df], ignore_index=True)
```
